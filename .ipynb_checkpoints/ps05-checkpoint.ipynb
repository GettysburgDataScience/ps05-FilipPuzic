### Problem 1

 - How many passengers are recorded in this log?
 - What percentage are male?
 - What percentage survived?

Hint: You can find these three answers with a single command.

````python
'''your code here'''

# Total passengers
total_passengers = len(titanic_data)

# Percentage of male passengers
male_percentage = (titanic_data['Sex'].sum() / total_passengers) * 100

# Percentage of passengers who survived
survival_percentage = (titanic_data['Survived'].mean()) * 100

total_passengers, male_percentage, survival_percentage
````

### Problem 2:

 - What percentage of female passengers survived?
 - What percentage of male passengers survived?

````python
'''your code here'''

# Filter the data by gender
female_passengers = titanic_data[titanic_data['Sex'] == 0]  # Assuming 0 = female
male_passengers = titanic_data[titanic_data['Sex'] == 1]    # Assuming 1 = male

# Calculate the survival rates for each gender
female_survival_rate = female_passengers['Survived'].mean() * 100
male_survival_rate = male_passengers['Survived'].mean() * 100

female_survival_rate, male_survival_rate
````

Your written answer:

 - <Answer 1> 74.20% of female passengers survived
 - <Answer 2> 18.9% of male passengers survived

### Problem 3

Use ```groupby``` and ```agg``` to find the percentage of the total tickets and the survivorship percentage for each class of ticket.

````python
'''your code here'''

# Total number of passengers
total_passengers = len(titanic_data)

# Group data by 'PClass' and calculate ticket percentages and survivorship percentages
ticket_survival_stats = titanic_data.groupby('Pclass').agg(
    ticket_percentage=('Pclass', lambda x: len(x) / total_passengers * 100),
    survivorship_percentage=('Survived', 'mean')
)

# Convert survivorship percentage to a percentage format
ticket_survival_stats['survivorship_percentage'] *= 100

ticket_survival_stats
````

Your written answer:

 - Class 1 
    - % tickets: 24.24
    - % survivorship: 62.96
- Class 2 
    - % tickets: 20.65
    - % survivorship: 47.28
- Class 3
    - % tickets: 55.1	
    - % survivorship: 24.24


### Problem 4

Create two new dataframes corresponding to survivors and non-survivors keeping only the columns: age, name, and sex. Display the first five rows of these new dataframes.

Create a figure showing the distribution of age by sex and survivorship (four histograms on the same axes: male-survivors, female-survivors, male-unlucky, female-unlucky). Choose colors and histtype (see the [documentation for histogram](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)) to make this figure easy to read and interpret.

````python
'''Your code here'''

# Create separate DataFrames for survivors and non-survivors, keeping only 'Age', 'Name', and 'Sex' columns
survivors = titanic_data[titanic_data['Survived'] == 1][['Age', 'Name', 'Sex']]
non_survivors = titanic_data[titanic_data['Survived'] == 0][['Age', 'Name', 'Sex']]

# Display the first five rows of each new DataFrame
print("Survivors:")
print(survivors.head())
print("Non-Survivors:")
print(non_survivors.head())

# Plotting the age distributions by sex and survivorship
plt.figure(figsize=(12, 8))

# Four histograms for male/female survivors/non-survivors
plt.hist(survivors[survivors['Sex'] == 1]['Age'].dropna(), bins=15, color='blue', alpha=0.5, label='Male Survivors', histtype='stepfilled')
plt.hist(survivors[survivors['Sex'] == 0]['Age'].dropna(), bins=15, color='green', alpha=0.5, label='Female Survivors', histtype='stepfilled')
plt.hist(non_survivors[non_survivors['Sex'] == 1]['Age'].dropna(), bins=15, color='red', alpha=0.5, label='Male Non-Survivors', histtype='stepfilled')
plt.hist(non_survivors[non_survivors['Sex'] == 0]['Age'].dropna(), bins=15, color='orange', alpha=0.5, label='Female Non-Survivors', histtype='stepfilled')

# Adding title, labels, and legend
plt.title("Age Distribution by Sex and Survivorship on the Titanic")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()

# Display the plot
plt.show()
````

### Problem 5

Now that we've inspected the data, let's build a classifier (you'll compare two models) that predicts which passengers survived. These are the steps you should take:

 - Choose your features. The label will be survivorship.
 - Split the data into training and test sets.
 - Standardize the values using StandardScaler
 - Choose two of the following model types: KNN, SVM, Decision Tree. Create those models and select hyper-parameters.
 - Fit your models to the training data.
 - Apply the fit models to create predictions from the test data.
 - For each model, create a confusion matrix for the training and test data. You may want to revisit the hyper-parameters and re-fit the models.

 - Answer the questions at the bottom of the code block.

````python
'''Your code here'''

features = ['Age', 'Sex', 'Pclass']
label = ['Survived']
````

````python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(titanic_data[features], titanic_data[label], test_size = 0.2)
````

````python
ss = StandardScaler()

ss.fit(x_train)

xs_train = ss.transform(x_train)
xs_test = ss.transform(x_test)
````

````python
from sklearn.tree import DecisionTreeClassifier, plot_tree

tree_clf = DecisionTreeClassifier(max_depth = 4, min_samples_leaf = 1)
tree_clf.fit(x_train, y_train)

plt.figure(figsize = (15,6))
plot_tree(tree_clf,
         class_names = ['drowned', 'survived'],
         feature_names = ['Age','Sex', 'Pclass'],
         fontsize = 10)

plt.show()
````

````python
from sklearn.metrics import ConfusionMatrixDisplay

y_tree_train = tree_clf.predict(x_train)
y_tree_test = tree_clf.predict(x_test)
````

````python
ConfusionMatrixDisplay.from_predictions(y_train, y_tree_train,
                                       normalize = 'true',
                                       cmap = 'Greens',
                                       display_labels = ['Did not','Survived'])
````

````python
ConfusionMatrixDisplay.from_predictions(y_test, y_tree_test,
                                       normalize = 'true',
                                       cmap = 'Greens',
                                       display_labels = ['Did not','Survived'])
````
